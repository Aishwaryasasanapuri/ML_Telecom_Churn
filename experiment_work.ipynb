```python
# Load clean telecom data file
master_df = pd.read_csv('telecom_churn_data_clean.csv', low_memory=False)
```

```python
master_df.head()
```

```python
# Splitting the data into train and test
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import f_regression
from sklearn.feature_selection import f_classif
X = master_df.drop(['churn'],axis=1)
y = master_df['churn']
feature=X.columns
selector_f= SelectPercentile(f_classif,percentile=25)
selector_f.fit(X,y)
print(len(selector_f.scores_))
for n,s in zip (feature,selector_f.scores_):
    print("importance of for feature %s",(s,n))
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_tr,y_tr = sm.fit_sample(X,y)
print(X_tr.shape)
print(y_tr.shape)
np.count_nonzero(y_tr)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tr,y_tr, train_size=0.7,test_size=0.3,random_state=100)
print(X_train.shape)
```
